```python
# میخاهیم به اشیا تابع و متد اضافه کنیم
# وقتی توابع درون کلاس تشکیل شوند به آن ها متد گفته میشود
# یک تفاوت بین متد و تابع وجود دارد و آن هم پارامتر self است

# self > instance variable
# یک کلمه قراردادی که همان خود آبجکت است که بصورت خودکار ارسال میشود
# وجود سلف اجباری است 

class Point:
    def reset(self):
        print(self)
        self.x = 0
        self.y = 0

p1 = Point()

p1.x = 4
p1.y = 2

p1.reset()

print(p1.x)
print(p1.y)

print()
# روش دیگر صدا زدن متد
Point.reset(p1)
```

```python
from math import hypot
# قاعده و ارتفاع را میدهیم وتر را محاسبه میکند


class Point:
    def move(self, x:float, y:float)->None:
        self.x = x
        self.y = y
        
    def reset(self):
        self.move(0, 0)
        
# فاصله دوتا نقطه
    def distance(self, other:"Point")->float:
        return hypot(self.x - other.x, self.y - other.y)

    
p1 = Point()
p2 = Point()

p1.reset()
p2.move(3, 4)

print(p1.distance(p2))
```

```python
# method
# attribute -> default / non default

# برای اینیت نباید ریترن نوشت

# در اینصورت باید شی هایی که میسازیم ورودی داشته باشند
class Point:
#     attribute non default
    # def __init__(self, x:float, y:float)->None::
    #     self.x = x
    #     self.y = y
    
#     attribute default
    # def __init__(self, x:float = 0, y:float = 0)->None::
    #     self.x = x
    #     self.y = y
    
#   methode  
    def __init__(self, x:float = 0, y:float = 0)->None:
        self.move(x, y)
    
    def move(self, x:float, y:float)->None:
        self.x = x
        self.y = y
        
    def reset(self):
        self.move(0, 0)
        
# فاصله دوتا نقطه
    def distance(self, other:"Point")->float:
        return hypot(self.x - other.x, self.y - other.y)

    
p1 = Point(0 , 0)
p2 = Point(3 , 4)

print(p1.distance(p2))
```

```python
class Point:
    """
    A point at the axis provides two-dimensional coordinates.
    >>> p1 = Point()
    >>> p2 = Point(3, 4)
    >>> p1.reset()
    >>> p1.distance(p2)
    5
    """
    def __init__(self, x:float = 0, y:float = 0)->None:
        """
        Intialize the x and y coordinates of the new point
        :param x: x-coordinates
        :param y: y-coordinates
        """
        self.move(x, y)
    
    def move(self, x:float, y:float)->None:
        self.x = x
        self.y = y
        
    def reset(self):
        self.move(0, 0)
        
# فاصله دوتا نقطه
    def distance(self, other:"Point")->float:
        return hypot(self.x - other.x, self.y - other.y)
   
p1 = Point(0 , 0)

# مشاهده داک استرینگ
print(Point.__doc__)

print(help(Point)) # this is better

# تست مثال هایی که زده اییم
import doctest
doctest.testmod()

```
