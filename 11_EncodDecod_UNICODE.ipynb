# Encoding and Decoding

انکودینگ : جایگزین کردن کاراکترها و رشته ها با کاراکتر ها و رشته هایی که انتقال و ذخیره سازی آن برای کامپیوتر بهینه باشد

# ASCII & UNICODE

همه چیز در حافظه به صورت بیت ها و بصورت صفر یا یک ذخیره می شود

اعداد در مبناهای دودویی قابل تبدیل هستند ولی کاراکتر ها و سیمبول ها قبل تبدیل نیستند

لذا یکسری اعداد به حرف ها تخصیص میدهیم. این کدهای عددی در استاندارد ها مختلفی هستند که اسکی و از جمله آنها است

یونیکد استاندارد دیکودینگی هست که برای استفاده از تمام زبانها و کاراکترها در دنیا تدوین شده است.

یونیکد یوتی اف 8 هست یونیکد یوتی اف 16 و 32 هم هست

زبان پایتون از یونیکد استفاده می کند و لذا از زبان فارسی نیز استفاده میشود

# متد های مربوط به یونیکد

```python
#ord() برگرداندن کد دسیمال یونیکد به عدد
ord("f")
```

```python
# bin(ord()) برگرداندن باینری بیت یونیکد
bin(ord("f"))
```

```python
# chr  کاراکتر مربوط به کد یونیکد
chr(1587)
```

```python
# برای چاپ کاراکتر های خاص با داشتن کد یو تی اف آن استفاده می کنیم
print("char : \u0489")  # utf8-chartable.de
print("char in octa : \052")  # نحوه انتخاب کاراکتر در مبنای 8
print("char in hex  : \x2a")  # نحوه انتخاب کاراکتر در مبنای 16
```

# نوع داده باینری Bytearray

برای عکس و فیلم و داده های شبکه مورد استفاده قرار می گیرد

```python
# python ver 2
# str اسکی و داده های باینری
# unicode رشته های یونیکد

# python ver 3
# str رشته های معمولی شامل اسکی و یونیکد
# bytes داده های باینری مربوط به بایت های متن انکود شده
# bytearray همان داده بایت ولی قابل تغییر یا ایمیوتیبل
```

```python
# نحوه بیان داده بایت
s = b"r"
print(s)
s = bytes("a", "utf-8")
print(s)
s = "tf".encode("utf-8")
print(s)
print(type(s))
```

```python
# bytearray
s = bytearray("abbas", "utf-8")
print(s)
```

```python
name = bytes('reza' , 'utf-8')
# \x72\x65\x7a\xc3\xa6

print(name[3])
print('\x72\x65\x7a\xc3\xa6')
# دکود بایت ها
print(name.decode('utf-8'))
print(str(name ,'utf-8'))
```

```python
name = bytes(8)
print(name)
```

```python
name = bytes([1, 3, 98, 56])
print(name)
```

```python
name = b'tf'
name[0] = 112
```

```python
name = bytearray('tf' , 'utf-8')
name[0] = 112
print(name.decode('utf-8'))
```

```python
name = bytes('tfبق' , 'ascii' , 'ignore')
name = bytes('tfبق' , 'ascii' , 'namereplace')

# strict ignore replace namereplace

print(name)
```

```python
print((5).to_bytes(4, 'big'))
print((32).to_bytes(4, 'big'))
print('\x20')
```
