## دستور with/as و شی context manager  <a class="anchor" id="section_1_6"></a> 

کانتکست منیجر دارای دو متد هستند اینتر و ایکزیت. شبیه دکوراتور قبل و بعد یکسری دستور کارهایی انجام می دهند. کانتکست منیجر ها با استفاده از کلاس تعریف و با ویز ساخت می شود

یکسری کانتکت منیجر های دیفالت در پایتون وجود دارد

```python
class A:
    def __enter__(self):
        print("start")
    def __exit__(self, exc_type, exc_val, exc_tb):
        print("end!")

        
with A():
    print("I am Reza!")
    print("Welcome!")
```

```python
# امکان استفاده از ریترن در تریع کانتکست منیجر
class A:
    def __enter__(self):
        print("start")
        return 10
    def __exit__(self, exc_type, exc_val, exc_tb):
        print("end!")

        
with A() as x:
    print("Welcome!")
    print("x :", x)
 
# کانتکست منیجر ها با ویث فعال میشود و قبل از دستور آنها اینتر خوانده میشود و بعد از پایان آنها اگزیت
# در بعد از ریترن شده اینتر را در ان قرار میدهد
```

```python
# فایل ها را می توان بصورت کانتکست منیجر باز کرد
```

```python
f = open("sample2.txt", "r")
# enter : فایل را باز می کند و فایل را برمیگرداند 
# exit : فایل را میبندد
```

```python
# wtih مطمئن می شود کا فایل بسته می شود 
# با کانتکست منیجر می توان منابع را بعد استفاده آزاد کرد
```

```python
with open("sample2.txt", "a") as f:
    f.write("\nwe test file opening with ContxMngr")
with open("sample2.txt", "r") as f:
    for line in f:
        print(line, end="")
```

```python
# with تو در تو
with open("sample2.txt", "r") as f1:
    with open("new.txt", "w") as f2:
        for line in f1:
            f2.write(line)
```

```python
from sys import stdin
# stdin -> keyboard .... input()
f = stdin.readline()
print(f)
```

```python
from sys import stdout
# stdout -> screen .... print()
stdout.write("Hello!")
```

```python
from sys import stderr
# stderr -> screen .... exc
stderr.write("Hello!")
```

```python
# بجای اسکرین در یک فایل نمایش میدهد
import sys
stdout_temp = sys.stdout
sys.stdout = open("new.txt", "w")
print("Hello World") # دیگر در صفحه نمایش نمی نویسد بلکه در فایل چاپ می کند
sys.stdout.close()
sys.stdout = stdout_temp 
```

```python
# خواندن از فایل
stdin_temp = sys.stdin
sys.stdin = open("sample2.txt", "r")
f = sys.stdin.readline() # از فایل می خواند
z = input()
print(z)
sys.stdin.close()
sys.stdin = stdin_temp 
```

```python
# آرگومان های print
# دو آرگومان دیگر غیر sep, end
# file = sys.stdout  (defualt)
# flush  فلاش کردن بافر
f = open("sample2.txt", "a")
print(1, 2, 3, 4, file=f, flush=True)
# به جای صفحه نمایش در فایل نگه می دارد
```

مهم ترین ویژگی فایل های جیسون این است که شکل داده را حفظ می کند. عدد را بصورت عدد و رشته را بصورت رشته منتقل میکند.

```python
with open("mycsv.csv", "w" , newline="") as cf:
    writer = csv.writer(cf)  # ساخت شیء رایتر
    writer.writerow(["id", "name", "lastname", "age"])
    data = [["1", "reza", "dolati", 26],
        ["2", "neda", "ahmadi", 20],
        ["3", "ali", "nosrati", 18]]
    writer.writerows(data)
    # for i in data:
    #     writer.writerow(i)
```
