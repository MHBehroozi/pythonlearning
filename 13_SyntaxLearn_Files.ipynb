بافر حافظه ای میانجی بین دو واحد مجزا است. مثلا پردازنده با سرعت بالا دستورات پرینت برای پرینتر با سرعت کم را در حافظه بافر قرار می دهد و پردازنده بخاطر سرعت کم پرینتر متوقف نمی شود. پرینتر با سرعت کم خود مستقلا دستورات را از بافر میخواند. بطور کلی بافر باعث افزایش سرعت می شود

```python
f = open("sample2.txt", "a+", encoding="utf-8")
f.write("\nReza Dolati ")
f.close()  # با این دستور فایل بسته می شود
# توصیه اکید میشود حتما در آخر فایل را ببندین 
```

    تا زمانی که اسکریپت تمام نشود یا فایل بسته نشود اطلاعاتی در فایل نوشته نمی شود.

```python
# flush
# خالی کردن بافر و درنتیجه نوشته شدن در فایل بیش از بسته شدن
f = open("sample2.txt", "a+", encoding="utf-8")
f.write("\nReza Dolati ")
f.flush()  # سریعا در فایل نوشته می شود
input("x: ")
f.close()
```

## دستور with/as و شی context manager  <a class="anchor" id="section_1_6"></a> 

کانتکست منیجر دارای دو متد هستند اینتر و ایکزیت. شبیه دکوراتور قبل و بعد یکسری دستور کارهایی انجام می دهند. کانتکست منیجر ها با استفاده از کلاس تعریف و با ویز ساخت می شود

```python
class A:
    def __enter__(self):
        print("start")
    def __exit__(self, exc_type, exc_val, exc_tb):
        print("end!")

        
with A():
    print("I am Reza!")
    print("Welcome!")
```

```python
# امکان استفاده از ریترن در تریع کانتکست منیجر
class A:
    def __enter__(self):
        print("start")
        return 10
    def __exit__(self, exc_type, exc_val, exc_tb):
        print("end!")

        
with A() as x:
    print("Welcome!")
    print("x :", x)
```

```python
# فایل ها را می توان بصورت کانتکست منیجر باز کرد
```

جیسون یک قالب استاندارد باز است که امکان تبادل داده ها در وب با استفاده از جفت های خصوصیت کلید را ممکن ساخته است. قرادادهای مورد استفاده ی جیسون برای تمامی برنامه نویسان از جمله برنامه نویس جاوا اسکریپت، پرل، پایتون، جاوا، سی پلاس پلاس و ... شناخته شده است

مهم ترین ویژگی فایل های جیسون این است که شکل داده را حفظ می کند. عدد را بصورت عدد و رشته را بصورت رشته منتقل میکند.

```python
# سبک وزن
# خواندن و نوشتن بسیار ساده تر
# مستقل از زبان
# قابل استفاده در اکثر زبان ها
```

```python
# در پایتون همه چی شیء هستند
# ولی لیست ها در جیسون به شکل آرایه منتقل می شوند
```

```python
## python | javascript
# int/float | Number
# True/False | true/false
# None | null
# dict | object
# list/tuple | array
# string | string
```

```python
# loads -> json string to python format
# load  -> file
# dumps -> python string to json format
# dump  -> file
```

```python
jn_str = "[1, 2, 3, 4]"
py_type = json.loads(jn_str)
print(jn_str, type(jn_str))
print(py_type, type(py_type))
```

```python
jn_str = '"[1, 2, 3, 4]"'
py_type = json.loads(jn_str)
print(jn_str, type(jn_str))
print(py_type, type(py_type))
```

```python
jn_str = 'null'
py_type = json.loads(jn_str)
print(jn_str, type(jn_str))
print(py_type, type(py_type))
```

```python
with open("myfile.json", "r") as jf:
    s = json.load(jf)

print(s)
print(type(s))
```

```python
d = None
j = json.dumps(d)
print(j)
print(type(j))
```

```python
with open("mycsv.csv", "w") as cf:
    writer = csv.writer(cf)  # ساخت شیء رایتر
    writer.writerow(["id", "name", "lastname", "age"])
    data = [["1", "reza", "dolati", 26],
        ["2", "neda", "ahmadi", 20],
        ["3", "ali", "nosrati", 18]]
    writer.writerows(data)
```
